```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 500)

y1 = np.sin(x) + 0.1 * np.random.randn(len(x))
y2 = np.cos(x) + 0.1 * np.random.randn(len(x))
trend = np.linspace(0, -0.5, len(x))
y3 = trend + 0.02 * np.random.randn(len(x))

offset = 1.5
y1_off = y1 + offset
y2_off = y2 + offset

fig, ax = plt.subplots(figsize=(8,3), facecolor='white')
ax.set_facecolor('white')
ax.plot(x, y1_off, linewidth=3)
ax.plot(x, y2_off, linewidth=3)
ax.plot(x, y3,        linewidth=3)

# fix axis limits
ax.set_xlim(0, 20)
ax.set_ylim(-1, 3)
ax.autoscale(False)

# styl
ax.set_xticks([])
ax.set_yticks([])
for s in ax.spines.values():
    s.set_linewidth(1.5)
ax.margins(0)
plt.tight_layout()
plt.savefig("./saved/multivariate_series_syn_real.png")
plt.show()
```
