```python
import pandas as pd
df = pd.read_json("../benchmark/data/real_data.json")
```

```python
import matplotlib.pyplot as plt
import numpy as np

value_counts = df.groupby("league", dropna=False).size()
value_counts.index = value_counts.index.to_series().fillna("null")
value_counts.plot(kind="bar", figsize=(6, 3))

# plt.xlabel("League", fontsize=20)
plt.xlabel("")
plt.ylabel("Count", fontsize=16)
plt.xticks(rotation=0, fontsize=14)
# plt.title("Entries per League", fontsize=18)
plt.tight_layout()
plt.savefig("./saved/leagues.png")
plt.show()
```
