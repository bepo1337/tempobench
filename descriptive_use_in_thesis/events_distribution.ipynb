```python
import pandas as pd
df = pd.read_json("../benchmark/data/real_data.json")
```

```python
reasons = ["regular interval", "transfer", "new coach", "injury end", "market value update"]

import matplotlib.pyplot as plt
from collections import Counter

reason_counts = Counter()
for entry in df["reason"]:
    parts = [r.strip() for r in entry.split(",")]
    for reason in parts:
        if reason in reasons:
            reason_counts[reason] += 1
        elif "injury" in reason and "injury end" not in reason:
            reason_counts["injury"] += 1

plt.figure(figsize=(10, 6))
plt.bar(reason_counts.keys(), reason_counts.values())
# plt.xlabel("Reason", fontsize=20)
plt.ylabel("Count", fontsize=16)
# plt.title("Distribution of sampled events", fontsize=18)
plt.xticks(rotation=25, fontsize=16)
plt.tight_layout()
plt.savefig("./saved/events_distribution.png")
plt.show()
```
